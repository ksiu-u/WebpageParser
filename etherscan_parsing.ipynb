```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def get_html_code(link):
    headers = {'User-Agent': "PostmanRuntime/7.32.3",
                'referer': 'https://www.google.com/'}
    try:
        res = requests.get(link, headers=headers)
        return res.text
    except:
        return 'RuntimeError: ошибка ссылки'
```

```python
with open ('websites.md') as file:
    data = file.readline()  # строка с заголовком
    data = file.readline()  # строка с знаками тире
    while 'Бизнес-информатика' not in data:
        data = file.readline()
    link = data.rstrip('\n')[1: -1].split('|')[2]
```

```python
soup = BeautifulSoup(get_html_code(link), 'html.parser')

crypto = soup.find_all('div', class_='hash-tag text-truncate fw-medium')

crypto_list = [elem.text for elem in crypto]
crypto_price = soup.find_all('div', class_='d-inline')
crypto_price_list = list(filter(lambda elem: elem.startswith('$'), [elem.text[1: -1].replace(',', '') for elem in crypto_price]))
crypto_and_price = dict(zip(crypto_list, crypto_price_list))
crypto_and_price_sorted = dict(sorted(crypto_and_price.items(), key=lambda elem: float(elem[1][1:]), reverse=True))
```

```python
# result = pd.DataFrame(list(crypto_and_price_sorted.items()), columns=['crypto', 'price'])
result = pd.DataFrame.from_dict(crypto_and_price_sorted, orient='index').reset_index()
result.columns = ['crypto', 'price']
```
