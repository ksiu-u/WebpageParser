```python
import requests
from IPython.display import clear_output
```

```python
def get_html_code(link):
    headers = {'User-Agent': "PostmanRuntime/7.32.3",
                'referer': 'https://www.google.com/'}
    try:
        res = requests.get(link, headers=headers)
        return res.text
    except:
        return 'RuntimeError: ошибка ссылки'
```

```python
with open ('websites.md') as file:
    data = file.readline()  # строка с заголовком
    data = file.readline()  # строка с знаками тире
    data = file.readline()
    while data:
        clear_output(wait=True)
        link = data.rstrip('\n')[1: -1].split('|')[2]
        name = data.rstrip('\n')[1: -1].split('|')[1]
        print(f'"{name}" {link}\n')
        print(get_html_code(link))
        print(f'\nУСПЕШНО: "{name}" {link}\n')
        data = file.readline()
        cont = int(input('Вывод следующей страницы? 1 - да, 2 - нет\n'))
        if cont == 1:
            continue
        else:
            print('Работа завершена')
            break
    print('Документ обработан')
```
